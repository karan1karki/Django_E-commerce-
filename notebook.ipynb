```python
# linear search implementation in Python

def linear_search(arr, target):
    for i, x in enumerate(arr):
        if x == target:
            return i
        return -1

arr = [10, 20, 30, 40, 50]
target = 30
result = linear_search(arr, target)
print(result)

# Time complexity: O(n)
```

```python
# Binary search implementation in Python

def binary_search(arr, target):
    low , high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr [mid] == target:
            return mid
        elif arr [mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
arr = [10, 20, 30, 40, 50]

target = 40
result = binary_search(arr, target)
print(result)

# Time complexity: O(log n)
```

```python
# Bubble Sort implementation in Python

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr


arr = [5, 4, 7, 1, 2, 3]
sorted_arr = bubble_sort(arr)
print(sorted_arr)
```

```python
#  selection sort implementation in Python

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr [min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

arr = [5, 4, 7, 1, 2, 3]
sorted_arr = selection_sort(arr)
print(sorted_arr)


# Time complexity: O(n^2)
```

```python
#  Linked List implementation in Python
    # --> single linked list

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

        
    # --> doubly linked list

class DoublyNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
    def insert(self, data):
        new_node = DoublyNode(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    # --> circular linked list

class CircularNode:
    def __init__(self, data):
        self.data = data
        self.next = None
class CircularLinkedList:
    def __init__(self):
        self.head = None
    def insert(self, data):
        new_node = CircularNode(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return
        last = self.head
        while last.next != self.head:
            last = last.next
        last.next = new_node
        new_node.next = self.head
```
